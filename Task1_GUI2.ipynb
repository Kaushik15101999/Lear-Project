```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from functools import partial 
from tkinter.ttk import *

import matplotlib
matplotlib.use('TkAgg')
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg, NavigationToolbar2Tk

import plotly.io as pio

import subprocess
import webbrowser
import sys

import tkinter as tk
from tkinter import *
from tkinter import filedialog

from PIL import ImageTk,Image

root=tk.Tk()
root.title('   Lear X Bar Run Chart')
root.geometry("1050x500")
root.iconbitmap(r"C:\\Users\HP\Desktop\Lear Intern\2_gui\lear_logo.ico")


#my_img = ImageTk.PhotoImage(Image.open("C:\\Users\HP\Desktop\Lear Intern\lear.png"))
#my_label = Label(image=my_img)
#my_label.pack()




def import_csv_data():
    global df1
    global v
    
    file_path = filedialog.askopenfilename()
    print(file_path)
    v.set(file_path)
    df1 = pd.read_csv (file_path,low_memory=False)
    df1['join']=" || "
    df1['d_and_t']=df1['Date'].astype(str)+df1['join'].astype(str)+df1[' Time'].astype(str)
    #print (df)

    


def drop_menu_s():
    #global row_s
    global v_s
    v_s=StringVar(root)
    options_s=list(df1['d_and_t'])
    v_s.set(options_s[0])
    popupMenu = Combobox(root, textvariable=v_s, values=options_s,width=45)
    Label(root,text='Choose start Time').grid(sticky=W,row=1,column=1,pady=5,padx=40)
    popupMenu.grid(row = 1, column =1,pady=5)
    #row_s=v_s.get()
    #print(row_s)
    
def drop_menu_e():
    #global row_e
    global v_e
    v_e=StringVar(root)
    options_e=list(df1['d_and_t'])
    v_e.set(options_e[0])
    popupMenu = Combobox(root, textvariable=v_e, values=options_e,width=45)
    Label(root,text='Choose end Time').grid(sticky=W,row=2,column=1,padx=40)
    popupMenu.grid(row = 2, column =1)
    #row_e=v_e.get()
    #print(row_e)
    
       
    
def drop_menu():
    global v_r
    #global inp
    global par
    v_r=StringVar(root)
    options=list(df.columns)
    v_r.set(options[0])
    popupMenu =Combobox(root, textvariable=v_r, values=options,width=45)
    Label(root,text='Choose the parameter').grid(sticky=W,row=3,column=1,padx=40)
    popupMenu.grid(row = 3, column =1)
    #par=v_r.get()
    #prine(par)


def selectrow():
    global row_s
    global row_e
    global row_start
    global row_end
    global df
    row_start=int(df1[df1['d_and_t']==v_s.get()].iloc[0,0])
    row_end=int(df1[df1['d_and_t']==v_e.get()].iloc[0,0])
    
    df = df1.iloc[int(row_start):int(row_end)+1].copy()
    print(row_start)
    print(row_end)
    #print(df1)
    #print(df)
    row_s=v_s.get()
    row_e=v_e.get()
    print(row_end-row_start)
    
def display_graph_1():
    inp=v_r.get()
    run_var = df[inp]
    var = inp 
    y_label = inp
    
    avg = run_var.mean()
    df['avg'] = avg
    
    std = run_var.std()
    
    target = avg
    tolerance = std
    upper_tol = target+tolerance
    lower_tol = target-tolerance
    
    #number = pd.Series(np.arange(df.shape[0]))
    number=df[' Time']
    
    df['target'] = target
    df['upper_tol'] = upper_tol
    df['lower_tol'] = lower_tol
    df['number'] = number
    
    upper_1 = avg + std*1
    upper_2 = avg + std*2
    upper_3 = avg + std*3

    lower_1 = avg - std*1
    lower_2 = avg - std*2
    lower_3 = avg - std*3

    df['upper_1'] = upper_1
    df['upper_2'] = upper_2
    df['upper_3'] = upper_3

    df['lower_1'] = lower_1
    df['lower_2'] = lower_2
    df['lower_3'] = lower_3

    fig1=plt.figure(figsize = (25,10))
               
    plt.plot(number,run_var, color='b', marker='o')
    plt.plot(number, df['avg'], linestyle='solid', linewidth=3, color='g')
    plt.plot(number, df['upper_1'], linestyle='dashed', linewidth=3, color='r')
    plt.plot(number, df['upper_2'], linestyle='dashed', linewidth=3, color='r')
    plt.plot(number, df['upper_3'], linestyle='dashed', linewidth=3, color='r')

    plt.plot(number, df['lower_1'], linestyle='dashed', linewidth=3, color='r')
    plt.plot(number, df['lower_2'], linestyle='dashed', linewidth=3, color='r')
    plt.plot(number, df['lower_3'], linestyle='dashed', linewidth=3, color='r')
    plt.xlabel('Time')
    plt.ylabel('{}'.format(y_label))
    plt.title('X-Bar Run Chart')
    plt.xticks(rotation=90)
    #entry2=tk.Entry(root,width=100, borderwidth=5,textvariable=inp).grid(row=3,column=1)
    return fig1


import chart_studio.plotly as py
#import plotly.graph_objs as go
import plotly.graph_objects as go
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import cufflinks as cf
init_notebook_mode(connected=True)
cf.go_offline()

def display_graph_2():
    pio.renderers.default='browser'
    
    inp=v_r.get()
    run_var = df[inp]
    var = inp 
    y_label = inp
    
    avg = run_var.mean()
    df['avg'] = avg
    
    std = run_var.std()
    
    target = avg
    tolerance = std
    upper_tol = target+tolerance
    lower_tol = target-tolerance
    
    number = pd.Series(np.arange(df.shape[0]))
    
    df['target'] = target
    df['upper_tol'] = upper_tol
    df['lower_tol'] = lower_tol
    df['number'] = number
    
    upper_1 = avg + std*1
    upper_2 = avg + std*2
    upper_3 = avg + std*3

    lower_1 = avg - std*1
    lower_2 = avg - std*2
    lower_3 = avg - std*3

    df['upper_1'] = upper_1
    df['upper_2'] = upper_2
    df['upper_3'] = upper_3

    df['lower_1'] = lower_1
    df['lower_2'] = lower_2
    df['lower_3'] = lower_3
    fig = go.Figure()
    t=df['d_and_t']
    fig.add_trace(go.Scatter(x=t,y=df['avg'],name="mean values"))
    fig.add_trace(go.Scatter(x=t,y=run_var,name="original values"))
    fig.add_trace(go.Scatter(x=t,y=df['upper_1'],name="upper_1"))
    fig.add_trace(go.Scatter(x=t,y=df['upper_2'],name="upper_2"))
    fig.add_trace(go.Scatter(x=t,y=df['upper_3'],name="upper_3"))

    fig.add_trace(go.Scatter(x=t,y=df['lower_1'],name="lower_1"))
    fig.add_trace(go.Scatter(x=t,y=df['lower_2'],name="lower_2"))
    fig.add_trace(go.Scatter(x=t,y=df['lower_3'],name="lower_3"))
    fig.update_layout(title="X-Bar Run Chart",xaxis_title="Day and Time",yaxis_title='{}'.format(y_label),
                  font=dict(family="Courier New, monospace",size=10,color="#7f7f7f"))
    #fig.show()
    return fig.show()



entry3=tk.Entry(root,width=100,font=('Helvetica', 10), borderwidth=5)
entry3.grid(row=8,column=1,padx=20)


entry4=tk.Entry(root,width=100,font=('Helvetica', 10), borderwidth=5)
entry4.grid(row=9,column=1)




def results():
    global res_lt
    inp=v_r.get()
    run_var = df[inp]
    var = inp 
    y_label = inp
    
    avg = run_var.mean()
    df['avg'] = avg
    
    std = run_var.std()
    
    target = avg
    tolerance = std
    upper_tol = target+tolerance
    lower_tol = target-tolerance
    
    number = pd.Series(np.arange(df.shape[0]))
    
    df['target'] = target
    df['upper_tol'] = upper_tol
    df['lower_tol'] = lower_tol
    df['number'] = number
    
    upper_1 = avg + std*1
    upper_2 = avg + std*2
    upper_3 = avg + std*3

    lower_1 = avg - std*1
    lower_2 = avg - std*2
    lower_3 = avg - std*3

    df['upper_1'] = upper_1
    df['upper_2'] = upper_2
    df['upper_3'] = upper_3

    df['lower_1'] = lower_1
    df['lower_2'] = lower_2
    df['lower_3'] = lower_3 
    
    errors=set()
    errors.clear
    differentials_to_target = run_var - target
    differentials_to_avg = run_var - avg
    first_differences = np.ediff1d(run_var)
    #tolerances
    if np.max(np.absolute(differentials_to_target)) > tolerance:
        errors.add("Dimensions are out of tolerance")
    #beyond limit
    absolute_differentials = np.absolute(differentials_to_avg)

    if np.max(absolute_differentials) > upper_3:
        errors.add("Outliers Exist")
    #outer region or Zone A
    for index, i in enumerate(absolute_differentials):
        if index < 2:
            continue
        count = np.count_nonzero(absolute_differentials[index-2:index+1] > (std*2))
        if count >= 2:
            errors.add("Outer Zone Clusters")
    #middle regions or Zone B
    for index, i in enumerate(absolute_differentials):
        if index < 4:
            continue
        count = np.count_nonzero(absolute_differentials[index-4:index+1] > (std*1))
        if count >= 4:
            errors.add("Middle Zone Clusters")
    #inner region or Zone C
    for index, i in enumerate(absolute_differentials):
        if index < 6:
            continue
        count = np.count_nonzero(differentials_to_avg[index-6:index+1] > 0)
        if count >= 7:
            errors.add("Inner Zone Clusters")
        count = np.count_nonzero(differentials_to_avg[index-6:index+1] < 0)
        if count >= 7:
            errors.add("Inner Zone Clusters")
    #trends
    for index, i in enumerate(first_differences):
        if index < 6:
            continue
    count = np.count_nonzero(first_differences[index-5:index+1] > 0)
    if count >= 7:
        errors.add("Trending Data is Present")
    count = np.count_nonzero(first_differences[index-5:index+1] < 0)
    if count >= 7:
        errors.add("Trending Data is Present")
    #mixture
    for index, i in enumerate(absolute_differentials):
        if index < 8:
            continue
        count = np.count_nonzero(absolute_differentials[index-8:index+1] > upper_3)
        if count == 0:
            errors.add("No Mixture")
        else:
            errors.add("Mixture")
    #stratification
    for index, i in enumerate(absolute_differentials):
        if index < 15:
            continue
        count = np.count_nonzero(absolute_differentials[index-15:index+1] > upper_1)
        if count == 0:
            errors.add("No Stratification")
    #over control
    def sign_change(x,y):
        if x > 0 and y > 0:
            return 0
        elif x < 0 and y < 0:
            return 0
        else:
            return 1 

    changes = []

    for index, i in enumerate(first_differences):
        if index == 0:
            continue
        change = sign_change(first_differences[index],first_differences[index-1])
        changes.append(change)

    for index, i in enumerate(changes):
        if index < 14:
            continue
        if np.array(changes[index-14:index+1]).sum() >= 15:
            errors.add("Over Control")
    
    
    entry3.insert(0,repr(errors))
    # Possible causes
    causes = set() 
    for ele in errors:
        if ele == "Outliers Exist" or ele == "Outer Zone Clusters":
            causes.add("New person doing the job, Wrong setup, Measurement error, Process step skipped, Process step not completed, Power failure, Equipment breakdown")
        elif ele == "Middle Zone Clusters" or ele == "Inner Zone Clusters":
            causes.add("Raw material change, Change in work instruction, Different measurement device/calibration, Different shift, Person gains greater skills in doing the job, Change in maintenance program, Change in setup procedure")  
        elif ele == "Trending data is present":
            causes.add("Tooling wear or Temperature effects (cooling, heating)") 
        elif ele == "Mixture" or ele == "Stratification":
            causes.add("More than one process present (e.g. shifts, machines, raw material.)") 
        elif ele == "Over Control":
            causes.add("Tampering by operator or Alternating raw materials") 
        else:
            causes.add("No error")
            
    entry4.insert(0,repr(causes))


def doNothing():
    print("nothing")
    
def create_window():
    window = tk.Tk() 
    
def graph_1():
    FigureCanvasTkAgg(display_graph_1(),root).get_tk_widget()
    
def graph_2():
    display_graph_2()
    


v=tk.StringVar()
par=tk.StringVar()
tk.Label(root,text='File Path').grid(row=0,column=0)



entry1 = tk.Entry(root, width=100, borderwidth=5,textvariable=v).grid(row=0, column=1)


#tk.Label(root,text='Results').grid(row=6,column=0)
#tk.Label(root, text="Reasons").grid(row=7,column=0)
#entry3=tk.Entry(root,textvariable=res_lt).grid(row=6,column=1)
#entry3.insert(0,results())
#entry4=tk.Entry(root,textvariable=).grid(row=7,column=1)

tk.Button(root,text='Results',command=results).grid(row=8,column=0,pady=20)
tk.Button(root,text='Possible Reasons',command=results).grid(row=9,column=0,pady=30)



tk.Button(root, text='Browse Data Set',
          command=import_csv_data).grid(row=0, column=2)

tk.Button(root, text='Close'
          ,command=root.destroy).grid(row=9, column=2)

tk.Button(root, text='Run Chart 1',
          command=graph_1).grid(row=5, column=0,padx=20)
tk.Button(root, text='Run Chart 2',
          command=graph_2).grid(row=6, column=0,pady=20)

tk.Button(root,text='Run Chart Parameters',command=drop_menu).grid(row=3,column=0,pady=20)
tk.Button(root,text='select rows',command=selectrow).grid(sticky=E,row=2,column=1,padx=50)

tk.Button(root,text='Start Time',command=drop_menu_s).grid(row=1,column=0,padx=20,pady=20)
tk.Button(root,text='End Time',command=drop_menu_e).grid(row=2,column=0)

root.columnconfigure(0,weight=1)
root.columnconfigure(1,weight=1)
root.columnconfigure(2,weight=1)

root.rowconfigure(0,weight=1)
root.rowconfigure(1,weight=1)
root.rowconfigure(2,weight=1)
root.rowconfigure(3,weight=1)
root.rowconfigure(4,weight=1)
root.rowconfigure(5,weight=1)
root.rowconfigure(6,weight=1)
root.rowconfigure(7,weight=1)
root.rowconfigure(8,weight=1)
root.rowconfigure(9,weight=1)



menu =  Menu(root)
root.config(menu=menu)
subMenu = Menu(menu)
menu.add_cascade(label="File",menu=subMenu)
subMenu.add_command(label="New", command=create_window)
subMenu.add_command(label="Open", command=doNothing)
subMenu.add_command(label="Restart", command=doNothing)
subMenu.add_command(label="Exit", command=doNothing)
editMenu = Menu(menu)
menu.add_cascade(label = "Help", menu=editMenu)
editMenu.add_command(label="Help", command=doNothing)

root.mainloop()
```
